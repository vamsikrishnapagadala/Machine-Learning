```python
mod = LabelEncoder()
mod.fit(data['Age'])
data['Age'] = mod.transform(data['Age'])
print(data['Age'].value_counts())
```

```python
data.head()
```

```python
mod1 = LabelEncoder()
mod1.fit(data['Year'])
data['Year'] = mod1.transform(data['Year'])
print(data.head())
```

```python
X = data[['Year','District.Code','Neighborhood.Code','Gender','Age']].values
y = data[['Number']]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.5,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
print(model.score(X_test,y_test))
```
